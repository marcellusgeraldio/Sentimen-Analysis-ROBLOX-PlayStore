```python
from google_play_scraper import app, reviews
import pandas as pd
import os
```

```python
# Menentukan ID aplikasi Gojek
app_id = 'com.roblox.client'

# Mengambil ulasan
reviews_data, _ = reviews(
    app_id,
    lang='id',
    country='id',
    count=10000
)

# Konversi ke DataFrame
reviews_df = pd.DataFrame(reviews_data)
reviews_df
```

```python
# simpan ke csv
reviews_df.to_csv('reviews_roblox.csv', index=False)
```
